```python
import requests
from lxml.html import fromstring
import pandas as pd
```

```python
response = requests.get('https://learning.oreilly.com/live-training/?upcoming=180&topic=')

doc = fromstring(response.text)

course_list = doc.find_class('course-list')[0]

xps = {
    'dates': './div[@class="coursemeta"]//time/span/time',
    'courses': './div[@class="coursemeta"]/div/a',
    'presenters': './div[@class="coursemeta"]/div/strong/span/span',
    'status': './div[@class="cta"]/div/div/span',
}
```

```python
def get_item(course,x):
    item = course.xpath(xps[x])
    if isinstance(item, list):
        item = ' '.join([_.text.strip() for _ in item])
    return item
```

```python
def get_item(course,x):
    item = course.xpath(xps[x])
    if isinstance(item, list):
        item = ' '.join([_.text.strip() for _ in item])
    return item
```

```python
dates = []
courses = []
presenters = [] 
status = []
for course in course_list:
    dates.append(get_item(course, 'dates'))
    courses.append(get_item(course, 'courses'))
    presenters.append(get_item(course, 'presenters'))
    status.append(get_item(course, 'status'))
```

```python
upcoming = pd.DataFrame({'Date':dates,
                        'Course':courses,
                        'Presenter':presenters,
                        'Status':status})
upcoming
```
